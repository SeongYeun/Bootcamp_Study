# UI 프로그래밍



- 픽셀 위치지정 : .pack()
- grid 위치지정 : .grid()
- pack()과 grid()는 하나의 윈도우창에 동시적용할 수 없음
- 텍스트 1줄 : .Label()
- 텍스트 1+줄 : .Text()



```python
import tkinter as tk
# from tkinter import *
```

```python
root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x200+1000+500")
# 윈도우 창 전체 크기(가로x세로) 및 창 위치(+가로위치+세로위치치) 설정 (방법 1)

# label = tk.Label(root, text="안녕하세요")
# label.pack(side="left")
# root.mainloop()             # 윈도우 프로그램 실행
```

* 레이아웃 (pack)
* x : 가로방향  /  y : 세로방향   /  padx : 가로 여백   /  pady : 세로 여백
* bg : background  /  fg : fontcolor



```python
label_1 = tk.Label(root, text="TOP", bg="red", fg="white")
label_1.pack(side="top", fill="x", padx=20, pady = 15)              # label의 위치 지정

label_2 = tk.Label(root, text="BOTTOM", bg="blue", fg="white")
label_2.pack(side="bottom", fill="x", padx=20, pady = 15)

label_3 = tk.Label(root, text="LEFT", bg="green", fg="white")
label_3.pack(side="left", fill="y", padx=20, pady = 15)

label_4 = tk.Label(root, text="RIGHT", bg="yellow", fg="black")
label_4.pack(side="right", fill="y", padx=20, pady = 15)

label_5 = tk.Label(root, text="CENTER", bg="purple", fg="white")
label_5.pack(side="top", fill="both", expand=True, padx=20, pady = 15)


root.mainloop()             # 윈도우 프로그램 실행
```

```python
# 레이아웃 ㅡ grid 방식
root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x200+1000+500")

label_1 = tk.Label(root, text="label_1", bg="red", fg="white")
label_1.grid(row=0, column=0, padx=10, pady=10)             # label의 위치 지정

label_2 = tk.Label(root, text="label_2", bg="blue", fg="white")
label_2.grid(row=30, column=10, padx=10, pady=10)

label_3 = tk.Label(root, text="label_3", bg="green", fg="white")
label_3.grid(row=1, column=10, padx=10, pady=10)

label_4 = tk.Label(root, text="label_4", bg="yellow", fg="black")
label_4.grid(row=1, column=1, padx=10, pady=10)

label_5 = tk.Label(root, text="label_5", bg="purple", fg="white")
label_5.grid(row=0, column=5, rowspan=2, columnspan=1, padx=10, pady=10)



root.mainloop()
```

```python
# 텍스트

root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x200+1000+500")

def show_text():
    entried = entry.get()
    label.config(text=f"입력된 문자는 : {entried}")
    entry.delete(0, tk.END)             # 엔트리에 입력된 문자열 삭제

entry = tk.Entry(root, width=30)
entry.pack()

button = tk.Button(root, text="버튼 클릭", command=show_text)
button.pack(side="left")

label = tk.Label(root, text="")
label.pack()

root.mainloop()
```

```python
# 텍스트 ㅡ 여러 줄

root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x200+1000+500")

def show_text():
    print(text.get("1.0", tk.END))      # 입력받은 전체 문자열 출력(터미널창에 출력)


text = tk.Text(root, width=40, height=10)
text.pack()

button = tk.Button(root, text="버튼 클릭", command=show_text)
button.pack(side="left")

label = tk.Label(root, text="")
label.pack()

root.mainloop()
```

```python
# 텍스트 ㅡ 여러 줄
root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x200+1000+500")

top_frame = tk.Frame(root, bg="lightblue")
top_frame.pack(fill='x')
tk.Label(top_frame, text="top frame").pack(pady=30)

bottom_frame = tk.Frame(root, bg="lightgreen")
bottom_frame.pack(fill='both', expand=True)
tk.Label(bottom_frame, text="bottom frame_ expand:True").pack()

root.mainloop()
```

```python
# 체크버튼 / 체크박스
root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x200+1000+500")

def show_state():
    print("check는 ", var.get())            # 체크 버튼을 클릭할때 마다 상태가 터미널창에 출력됨


var = tk.IntVar()
checkbtn = tk.Checkbutton(root, text="동의합니다.", variable=var, command=show_state)
checkbtn.pack()

root.mainloop()
```

```python
# radio button
root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x200+1000+500")

def show_choice():
    print("선택 : ", var.get())

var = tk.StringVar(value="option_1")            # value="option_1" : 창이 처음 열렸을때 선택되어있는 기본값 (없으면 둘다 선택되어짐)
radio_1 = tk.Radiobutton(root, text="옵션1", variable=var, value="option_1", command=show_choice)
radio_1.pack(pady=10)

radio_2 = tk.Radiobutton(root, text="옵션2", variable=var, value="option_2", command=show_choice)
radio_2.pack(pady=10)

root.mainloop()
```

```python
# 메뉴
from tkinter import messagebox

root = tk.Tk()
root.title("윈도우 프로그래밍 연습")
root.geometry("400x300+1000+500")

def new_file():
    messagebox.showinfo("메뉴", "파일이 선택되었습니다.")

def exit_app():
    root.quit()             # 윈도우 창 종료

# 메뉴바 만들기
menubar = tk.Menu(root)

filemenu = tk.Menu(menubar, tearoff=0)          # tearoff=0 : 비활성화 = 서브메뉴 독립적인 창 생성 안함
filemenu.add_command(label="New", command=new_file)
filemenu.add_separator()
filemenu.add_command(label="Exit", command=exit_app)

menubar.add_cascade(label="파일", menu=filemenu)
root.config(menu=menubar)                       # 메뉴옵션 설정(config)


root.mainloop()
```
