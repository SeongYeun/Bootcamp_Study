```python
label_1 = tk.Label(root, text="TOP", bg="red", fg="white")
label_1.pack(side="top", fill="x", padx=20, pady = 15)              # label  

label_2 = tk.Label(root, text="BOTTOM", bg="blue", fg="white")
label_2.pack(side="bottom", fill="x", padx=20, pady = 15)

label_3 = tk.Label(root, text="LEFT", bg="green", fg="white")
label_3.pack(side="left", fill="y", padx=20, pady = 15)

label_4 = tk.Label(root, text="RIGHT", bg="yellow", fg="black")
label_4.pack(side="right", fill="y", padx=20, pady = 15)

label_5 = tk.Label(root, text="CENTER", bg="purple", fg="white")
label_5.pack(side="top", fill="both", expand=True, padx=20, pady = 15)


root.mainloop()             #   
```

```python
#   grid 
root = tk.Tk()
root.title("  ")
root.geometry("400x200+1000+500")

label_1 = tk.Label(root, text="label_1", bg="red", fg="white")
label_1.grid(row=0, column=0, padx=10, pady=10)             # label  

label_2 = tk.Label(root, text="label_2", bg="blue", fg="white")
label_2.grid(row=30, column=10, padx=10, pady=10)

label_3 = tk.Label(root, text="label_3", bg="green", fg="white")
label_3.grid(row=1, column=10, padx=10, pady=10)

label_4 = tk.Label(root, text="label_4", bg="yellow", fg="black")
label_4.grid(row=1, column=1, padx=10, pady=10)

label_5 = tk.Label(root, text="label_5", bg="purple", fg="white")
label_5.grid(row=0, column=5, rowspan=2, columnspan=1, padx=10, pady=10)



root.mainloop()
```

```python
# 

root = tk.Tk()
root.title("  ")
root.geometry("400x200+1000+500")

def show_text():
    entried = entry.get()
    label.config(text=f"  : {entried}")
    entry.delete(0, tk.END)             #    

entry = tk.Entry(root, width=30)
entry.pack()

button = tk.Button(root, text=" ", command=show_text)
button.pack(side="left")

label = tk.Label(root, text="")
label.pack()

root.mainloop()
```

```python
#    

root = tk.Tk()
root.title("  ")
root.geometry("400x200+1000+500")

def show_text():
    print(text.get("1.0", tk.END))      #    ( )


text = tk.Text(root, width=40, height=10)
text.pack()

button = tk.Button(root, text=" ", command=show_text)
button.pack(side="left")

label = tk.Label(root, text="")
label.pack()

root.mainloop()
```

```python
#    
root = tk.Tk()
root.title("  ")
root.geometry("400x200+1000+500")

top_frame = tk.Frame(root, bg="lightblue")
top_frame.pack(fill='x')
tk.Label(top_frame, text="top frame").pack(pady=30)

bottom_frame = tk.Frame(root, bg="lightgreen")
bottom_frame.pack(fill='both', expand=True)
tk.Label(bottom_frame, text="bottom frame_ expand:True").pack()

root.mainloop()
```

```python
#  / 
root = tk.Tk()
root.title("  ")
root.geometry("400x200+1000+500")

def show_state():
    print("check ", var.get())            #       


var = tk.IntVar()
checkbtn = tk.Checkbutton(root, text=".", variable=var, command=show_state)
checkbtn.pack()

root.mainloop()
```

```python
# radio button
root = tk.Tk()
root.title("  ")
root.geometry("400x200+1000+500")

def show_choice():
    print(" : ", var.get())

var = tk.StringVar(value="option_1")            # value="option_1" :      (  )
radio_1 = tk.Radiobutton(root, text="1", variable=var, value="option_1", command=show_choice)
radio_1.pack(pady=10)

radio_2 = tk.Radiobutton(root, text="2", variable=var, value="option_2", command=show_choice)
radio_2.pack(pady=10)

root.mainloop()
```

```python
# list box
root = tk.Tk()
root.title("  ")
root.geometry("400x300+1000+500")

def show_selected():
    selection = listbox.curselection()          #    (curselection()) selection 
    print("selection   : ", selection)
    if selection:
        print(f"  : {listbox.get(selection[0])}")


listbox = tk.Listbox(root)
listbox.pack(pady=10)

for item in ["", "", "", ""]:
    listbox.insert(tk.END, item)

button = tk.Button(root, text="", command=show_selected)
button.pack(pady=10)

root.mainloop()
```

```python
# 
from tkinter import messagebox

root = tk.Tk()
root.title("  ")
root.geometry("400x300+1000+500")

def new_file():
    messagebox.showinfo("", " .")

def exit_app():
    root.quit()             #   

#  
menubar = tk.Menu(root)

filemenu = tk.Menu(menubar, tearoff=0)          # tearoff=0 :  =     
filemenu.add_command(label="New", command=new_file)
filemenu.add_separator()
filemenu.add_command(label="Exit", command=exit_app)

menubar.add_cascade(label="", menu=filemenu)
root.config(menu=menubar)                       #  (config)


root.mainloop()
```
