```python
import pandas as pd
import numpy as np

# 예제 데이터프레임 생성
df_with_na = pd.DataFrame({
    'A': [1, 2, np.nan, 4, np.nan],
    'B': [np.nan, 2, 3, 4, np.nan],
    'C': [1, np.nan, np.nan, 4, np.nan],
    'D': [np.nan, np.nan, np.nan, np.nan, np.nan],
    'E' : [1, 2, 3, 4, 5]
})
df_with_na
```

```python

# axis=1 방향
df_axis_1_0 = df_with_na.dropna(thresh=3, axis=1)
df_axis_1_1 = df_with_na.dropna(how='all', axis=1)
df_axis_1_2 = df_with_na.dropna(how='any', axis=1)
result = pd.concat([df_with_na, df_axis_1_0, df_axis_1_1, df_axis_1_2], 
                    keys=["Origianl","thresh=3", "how=all", "how=any"],
                    axis=1)
result
```

```python
# axis=0 방향
df_axis_1_0 = df_with_na.dropna(how='any', subset=["B", "C"])
df_axis_1_1 = df_with_na.dropna(how='all', subset=["B", "C", "D"])
result = pd.concat([df_with_na, df_axis_1_0, df_axis_1_1],
                    keys=["Origianl","how=any in B~C", "how=all in B~D"])
result
```

```python
df_with_na
```

```python
mean_axis_0 = pd.DataFrame(df_with_na.mean()).T
fill_axis_0 = df_with_na.apply(lambda col : col.fillna(col.mean()))
fill_ffill = df_with_na.ffill()     # fillna(method='ffill') 문법이 .ffill()메서드로 대체 예정정
results = pd.concat([df_with_na, mean_axis_0, fill_axis_0, fill_ffill], 
                    keys = ["Original", "mean", "filled w/ mean", "ffill"])
results
```

```python
# axis=1방향 fillna()
median_axis_1 = pd.DataFrame(df_with_na.median(axis=1), columns=['median'])
fill_axis_1 = df_with_na.apply(lambda row : row.fillna(median_axis_1.loc[row.name, 'median']), axis=1)
fill_bfill = df_with_na.bfill()     # fillna(method='bfill') 문법이 .bfill()메서드로 대체 예정
results = pd.concat([df_with_na, median_axis_1, fill_axis_1, fill_bfill], axis=1,
                    keys = ["Original", "median", "filled w/ median", "bfill"])
results
```

```python
# programmers : 숫자 짝꿍
def solution(X, Y):
    answer = ''
    result = []
    X = [int(i) for i in list(X)]
    Y = [int(i) for i in list(Y)]
    if len(X)<=len(Y):
        short, long = X, Y
    else : 
        short, long = Y, X
    # print(short, long)
    
    for i in range(len(short)):
        c = short[i]
        if c in long :
            result = result.extend([c])
            long.remove(c)
            # result = result.append(c)
            print(c, result)
    
    result = result.sort(reverse=True)
    print(result)
    answer = str(int("".join(result)))
    if not result :
        answer = '-1'
    return answer

solution("23486", "2345")
```
