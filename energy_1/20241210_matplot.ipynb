※ 히스토그램 (histogram)

```python
import matplotlib.pyplot as plt

# 한글 설정
# 기본적으로 matplotlib은 한글 지원이 안되서 설정안하면 한글이 깨짐
from matplotlib import font_manager
path =  'C:\\Windows\\Fonts\\LG PC.ttf'
font = font_manager.FontProperties(fname=path).get_name()
plt.rc('font', family=font)
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 5, 6]
plt.hist(data)
# bins : 기본값 10 (구간갯수)  => (max-min)/bins
```

```python
# bins=5 일때
plt.hist(data, bins=5, color="orange", histtype="step")
plt.title("히스토그램", fontsize=15, pad=20)
plt.xlabel("값", fontsize=12)
plt.ylabel("빈도수", fontsize=12)
```

```python
# 히스토그램 ㅡ 데이터 2표본일때
data_1 = [1, 2, 2, 3, 3, 3, 4]
data_2 = [2, 3, 3, 4, 4, 5, 6]

plt.hist([data_1, data_2], bins=5, color=["g", "b"], alpha=0.4, label=['data_1', 'data_2'])
# alpha : 1에 가까울수록 짙음 (= 투명도 0)
plt.title("2표본 히스토그램")
plt.xlabel("값")
plt.ylabel("빈도수")
plt.legend()
```

※ 산점도 (Scatter)

```python
x = list(range(1, 5+1))
y = [2, 3, 5, 7, 11]
size = [20, 50, 80, 100, 200]
colors=list(range(10, 50+1, 10))

plt.scatter(x,y, s=size, c=colors, cmap="viridis")
plt.colorbar(label="color_bar")
```

```python
import numpy as np

n=50
x=np.random.rand(n)         # 난수 생성
y=np.random.rand(n)         # 난수 생성

area = (np.random.rand(n)*30) **2
colors = np.random.rand(n)

plt.scatter(x,y, s=area, c=colors, cmap="Spectral", alpha=0.7)
plt.colorbar(label="color_bar")
```

※ 파이차트(Pie)

```python
sizes = [25, 25, 20, 20]
label = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=label)
```

```python
sizes = [15, 30, 45, 10]
label = ['사과', '바나나', '체리', '멜론']
exploding = [0, 0, 0.1, 0]         # 강조할 범주만 0이 아닌 값으로 리스트 생성
plt.pie(sizes, labels=label, explode=exploding, autopct="%1.1f%%", shadow=True, startangle=0)
```

```python
sizes = list(range(10, 40+1, 10))
label = ['A', 'B', 'C', 'D']
color = ['c', 'm', 'y', 'g']

plt.pie(sizes, 
        labels=label, 
        colors = color,
        autopct="%1.1f%%", 
        textprops={"fontsize":20, "color":"darkblue"}, 
        wedgeprops={"edgecolor":"k"})
```

```python
# 실습1. 그래프 그리기
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.figure(figsize=(10, 6))             # 출력 창의 크기 설정 (인치 기준)
plt.plot(months, sales_2019, alpha=0.7, label="2019")
plt.plot(months, sales_2020, color="orange", label="2020")
plt.title("Monthly Sales Comparison (2019-202)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
```

```python
# 실습2. 그래프 그리기
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data=[20,35,15,27,45]
plt.bar(categories, data)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(True, axis="both", linestyle="-", color="lightgray")
plt.ylim([0, 50])
plt.xticks(rotation=45)
```

```python
# 실습3. 그래프 그리기
label = ['Apple', 'Banana', 'Melon', 'Grapes']
sizes = [0.34, 0.32, 0.16, 0.18]
color = ['r', 'y', 'g', 'purple']
expld = [0, 0.1, 0, 0.07]
plt.pie(sizes, 
        labels=label, 
        colors = color,
        autopct="%1.1f%%", 
        explode=expld,
        textprops={"fontsize":12, "color":"black"}, 
        wedgeprops={"width":0.7, "edgecolor":"k"})
```
