matplotlib 기본 설정 (한글 폰트)

```python
import matplotlib.pyplot as plt
```

```python
# 한글 설정
# 기본적으로 matplotlib은 한글 지원이 안되서 설정안하면 한글이 깨짐
from matplotlib import font_manager

# 1) 폰트 경로 찾기 ㅡ> 출력된 목록에서 원하는 폰트의 전체 경로를 복사
font_list = font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
font_list       # 설치되어 있는 폰트 목록
```

```python
path =  'C:\\Windows\\Fonts\\LG PC.ttf'
font = font_manager.FontProperties(fname=path).get_name()
plt.rc('font', family=font)
```

1) 기본 그래프 그리기

```python
# 선형 그래프 그리기
x = range(1,6)
y = range(2, 11, 2)
# 그래프 제목 ㅡ .title()
plt.title("맷플롯립_matplotlib")
# 그래프 생성 ㅡ .plot(x_df, y_df, ~ )
plt.plot(x, y, color = "purple", linestyle="--", linewidth=3)
```

```python
# 범례 ㅡ .legend()
plt.plot(x, y, color = "purple", linestyle="--", linewidth=3, label="샘플 그래프")
plt.legend(title="범주구분", fontsize=13, loc="center right")
# loc(범례위치) : "lower / cencter / upper" + "left / center / right"
plt.title("맷플롯립_matplotlib")
```

```python
# 마커(표식) ㅡ marker / markersize / markerfacecolor / markeredgecolor
plt.plot(x, y, marker="s", markersize=10, markerfacecolor="yellow", markeredgecolor="red", label="범주_1")
plt.legend(title="범주구분", fontsize=13, loc="center right")
plt.title("맷플롯립_matplotlib")
```

```python
# 제목(title 옵션설정) ㅡ fontdict 사용
keys_li = ["size", "color", "backgroundcolor", "weight"]
values_li = [20, "white", "black", "bold"]
font = dict(zip(keys_li, values_li))

plt.plot(x, y, marker="*", markersize=10, markerfacecolor="yellow", markeredgecolor="red", label="범주_1")
plt.legend(title="범주구분", fontsize=13, loc="center right")
plt.title("맷플롯립_matplotlib", pad=10, fontdict=font)
```

```python
# X/Y축 레이블 설정 ㅡ .xlabel  .ylabel
plt.plot(x, y, marker="*", markersize=10, markerfacecolor="yellow", markeredgecolor="red", label="범주_1")
plt.legend(title="범주구분", fontsize=13, loc="center right")
plt.title("맷플롯립_matplotlib", pad=10, fontdict=font)
plt.xlabel("X_축", labelpad=10, fontsize=15, color="green")
plt.ylabel("Y_축", fontsize=15, color="green")
```

```python
# 눈금선(격자) 옵션 ㅡ .grid
plt.plot(x, y, marker="*", markersize=10, markerfacecolor="yellow", markeredgecolor="red", label="범주_1")
plt.legend(title="범주구분", fontsize=13, loc="center right")
plt.grid(True, axis="both", linestyle="--", color="blue", alpha=0.2)
# axis = "X / Y / both"
# alpha (투명도)
plt.title("맷플롯립_matplotlib", pad=10, fontdict=font)
plt.xlabel("X_축", labelpad=10, fontsize=15, color="green")
plt.ylabel("Y_축", fontsize=15, color="green")
```

```python
# X/Y축 범위 지정 ㅡ .xlim  .ylim
plt.plot(x, y, marker="*", markersize=10, markerfacecolor="yellow", markeredgecolor="red", label="범주_1")
plt.legend(title="범주구분", fontsize=13, loc="center right")
plt.grid(True, axis="both", linestyle="--", color="blue", alpha=0.2)
plt.title("맷플롯립_matplotlib", pad=10, fontdict=font)
plt.xlabel("X_축", labelpad=10, fontsize=15, color="green")
plt.ylabel("Y_축", fontsize=15, color="green")

# 1) 축 각각 지정
# plt.xlim([0, 10])
# plt.ylim([0, 15])
# 2) x축과 y축을 동일하게
plt.axis("scaled")
# equal : 축 범위 동일하게
# scaled : x축과 y축의 스케일을 동일한 상태에서 화면에 거의 차도록 
# tight : 화면에 거의 차도록 축 범위와 스케일 자동 설정
# auto : 알아서 축 범위 설정
# [x_min, x_max, y_min, y_max] : 리스트로 각 축 범위를 직접 설정 가능
```

```python
# 여러 그래프 동시 그리기
import numpy as np
x = np.array([1, 2, 3, 4])
y1 = x
y2 = y1*2
y3 = y1*3
y4 = y1*4

plt.plot(x, y1, label = "y1=x", color = "r", marker="o")
plt.plot(x, y2, label = "y1=2x", color = "g", marker="*")
plt.plot(x, y3, label = "y1=3x", color = "b", marker=">")
plt.plot(x, y4, label = "y1=4x", color = "k", marker="<")

plt.legend(title="범주구분", fontsize=8, loc="upper left", ncol=4)

plt.title("여러 그래프 동시 그리기", pad=10, fontdict=font)
plt.xlabel("X_축", labelpad=10, fontsize=15, color="green")
plt.ylabel("Y_축", fontsize=15, color="green")
```

```python
# 여러 title의 여러 그래프 동시 출력 ㅡ .subplot(nrow, ncols, index)
plt.subplot(2, 2, 1)
plt.plot(x,y1)
plt.title("y1=x")

plt.subplot(2, 2, 2)
plt.plot(x,y2)
plt.title("y2=x")

plt.subplot(2, 2, 3)
plt.plot(x,y3)
plt.title("y3=x")

plt.subplot(2, 2, 4)
plt.plot(x,y4)
plt.title("y4=x")

plt.suptitle("2x2 그래프 그리기", fontsize=20)
plt.tight_layout()
```

※ 막대그래프 그리기 ㅡ .bar(x_df, y_df, ~)

```python
categories = ['A', 'B', 'C']
values = [10, 15, 7]

v_bars = plt.bar(categories, values, width=0.5, color=["r", "g", "b"], alpha=0.5, edgecolor="k", linewidth=1.5)
# width = 막대의 좌우 폭 너비
for bar in v_bars:
    plt.text( bar.get_x() + bar.get_width() / 2,        # : X축의 중심
              bar.get_height() + 0.7,                   # : 범주별 Y값에 0.5만큼 위
               str(bar.get_height()),                   # : 표시할 문구 (y값을 문자로 전환)
                ha="center", va="top", color="m" )      # : 각 bar(막대)를 기준으로 위치와 색상 설정

# 막대별(범주별) 이름 지정 ㅡ .xticks(x_df, [이름 목록])
plt.xticks(categories, ['2023', '2024', '2025'])
plt.title("막대그래프", pad=20, fontsize=17)
plt.xlabel("카테고리")
plt.ylabel("수치")
```

```python
categories = ['A', 'B', 'C']
values = [10, 15, 7]

h_bars = plt.barh(categories, values, color=["c", "m", "k"], alpha=0.7, edgecolor="r", linewidth=1.5)
# width = 막대의 좌우 폭 너비
for bar in h_bars:
    plt.text( bar.get_width() + 0.7,                    # : x축 값에 0.7만큼 우측 위치
              bar.get_y() + bar.get_height() / 2,       # : Y축의 중심
               str(bar.get_width()),                    # : 표시할 문구 (y값을 문자로 전환)
                ha="right", va="center", color="k" )    # : 각 bar(막대)를 기준으로 위치와 색상 설정

# 막대별(범주별) 이름 지정 ㅡ .xticks(x_df, [이름 목록])
# plt.xticks(categories, ['2023', '2024', '2025'])
plt.title("막대그래프", pad=20, fontsize=17)
plt.xlabel("카테고리")
plt.ylabel("수치")

```
